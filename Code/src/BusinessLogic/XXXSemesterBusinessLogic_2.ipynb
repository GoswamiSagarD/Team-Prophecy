# Business Problem Definition #


## Students ##

- Each Student pertains to one unique student
- Students can either be WAITLISTED or REGISTERED
- Students can select multiple courses, but can only choose ONE course section
- Students CANNOT retake a course they've already taken
- Students have a certain number of credits/courses they must take
- Some variability is required, so here is what we're going to do:
    + We must take the ratio of the current registration list for our subset and apply it to all REGISTERED students
    + Waitlisted students allow for an expansion of the total capacity.

## Courses ##
- Each Course contains a certain number of WAITLISTED students
- Each Course contains a certain number of REGISTERED students
- It's ok if a course is MISSING REGISTERED students
- Sections are smaller partitions of a course, but are UNIQUE
- Courses can be a CORE or ELECTIVE course. CORE courses are listed as required.
- Courses that ARE passed cannot be re-taken (our assumption). 
    + This is dependent on whether at ANY point in time the student takes the course again throughout the data.
- The SIZE of the course is not known. We are trying to FIND the size by determining the AVERAGE or MEDIAN of the number of students available.

## Sections ##
- Each Course Section is specific to ONE course
- ONE student can apply to ONE section
- Each section cannot admit students who have already taken the course
    + This can be done by reviewing the course.

## Program ##
- Each program contains a certain number of COURSES
- Each program has a certain influx of students that is distributed across each COURSE
- Each program is paired with a specific semester
- Each program can be paired based on prior semester performance
    + Programs that are compared against prior semesters will be compared against their REGISTERED and WAITLISTED students.
    + This extends down into COURSES.

## Complex Constraints to be implemented ##


**Indices**

$ s\in SemesterProgram $ = [1..N]

$c\in Course$ = [1..N]

$z\in Course Section$ = [1..N]

$d\in Student ID$ = [1..N]


_Only for Student IDs_
$v\in Visa$ = [0..1]

_Only for Courses and Course Section_
$r\in Required$ = [0..1]


**Indexed Variables**

$CAP_s$ = Median/Average number of expected students who will be taking the course. (Based on which is larger)

$PCS_{sc}$ = The ratio of students expected to take this course. _This is based on what our semester's number of registered students were previously_

$O_{scd}$ = The MAXIMUM number of courses a student can take in a semester

$P_{scd}$ = The MINIMUM number of required courses a student must take

**Decision Variables**

$ X_{sczdv} $ = Students who successfully registered

$ W_{sczdv} $ = Students who were waitlisted

$ A_{scr} $ = This counts the total number of students who registered for a class

$ B_{scr} $ = This counts the total number of students who were waitlisted for a class

$ S_{scdr} = \left\{
    \begin{array}\\
        1 & \mbox{if the student hasn't taken the course} \  \\
        0 & \mbox{else}
    \end{array}
\right. $

**Optimization Function**

$$ \text{Min} \sum_{\substack{s\in Semester\\c\in Course\\z\in Course Section\\d\in StudentID\\v\in Visa}}^{} W_{sczdv} $$


**Constraints**

**v---COMPLETED---v**

This weights students based on who is in the class. _This is effectively a flow diagram_:

$$ PCS_{scr} * A_{scr} <= O_{scr} + B_{scr} \text{ ; } \forall{s\in Semester\\c\in Course} $$

We must have a maximum number of classes the student takes in place. _This helps ensure diversity_:

$$ \sum_{\substack{s\in Semester\\c\in Course\\z\in Course Section}}^{} X_{sczdv} <= O_{scd} \text{ ; } \forall{d\in StudentID} $$

We must have a minimmum number of classes the student takes in place. _This helps ensure diversity_:

$$ \sum_{\substack{s\in Semester\\c\in Course\\z\in Course Section\\v\in Visa}}^{} (X_{sczdv} * S_{scd1}) >= P_{scd} \text{ ; } \forall{d\in StudentID}  $$

**v---COMPLETED---v**

If only $ X_{sczdv} $ or $ W_{sczdv} $ can exist at any given time. _They cannot both be used_:

$$ \sum_{\substack{v\in Visa}}^{} X_{sczdv} + \sum_{\substack{v\in Visa}}^{} W_{sczdv} <= 1 \text{ ; } \forall{s\in Semester\\c\in Course\\z\in Course Section \\d\in StudentID} $$

Students can only take ONE class per class section:

$$ \sum_{\substack{s\in Semester\\z\in Course Section\\v\in Visa}}^{} X_{sczdv} <= 1 * (1-S_{cdr}) \text{ ; } \forall{c\in Course \\d\in StudentID} $$

$$ \sum_{\substack{s\in Semester\\z\in Course Section\\v\in Visa}}^{} W_{sczdv} <= 1 * (1-S_{cdr}) \text{ ; } \forall{c\in Course \\d\in StudentID} $$

This maps all students under a registered class to our $ A_{sc} $ and $ B_{sc} $ values effectively:

$$ \sum_{\substack{z\in Course Section\\d\in StudentID\\v\in Visa}}^{} X_{sczdv}  =  A_{scr} \text{ ; } \forall{s\in Semester \\c\in Course } $$

$$ \sum_{\substack{z\in Course Section\\d\in StudentID\\v\in Visa}}^{} W_{sczdv} = B_{scr} \text{ ; } \forall{s\in Semester \\c\in Course } $$


Our $ X_{sczd} $ or $ W_{sczd} $ values will be binary:

$$ X_{sczd} = \left\{
    \begin{array}\\
        1 & \mbox{if the student is registered} \  \\
        0 & \mbox{else}
    \end{array}
\right. $$

$$ W_{sczd} = \left\{
    \begin{array}\\
        1 & \mbox{if the student is waitlisted} \  \\
        0 & \mbox{else}
    \end{array}
\right. $$
