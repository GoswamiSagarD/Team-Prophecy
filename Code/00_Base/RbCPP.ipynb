## Exporting the data to CSV

```python
# Get Course Enrollment HIstory for each course in each registration semester and student admit term
df = db_FinalSnapshot.runQuery(""" --sql
    SELECT stu_prog_desc, stu_visa, reg_status, stu_admit_term_desc, reg_term_desc, crs, COUNT(*) AS demand
    FROM FinalSnapshot
    GROUP BY
        stu_admit_term_desc, reg_term_desc, crs
    ORDER BY
        stu_admit_term_code, reg_term_code, crs
""")

df.to_csv( os.path.join('Data', '02_processed', 'RbCPP', 'FinalSnapshot.csv'), index=False)
df.to_pickle( os.path.join('Data', '02_processed', 'RbCPP', 'FinalSnapshot.pkl'))
# Exporting to SQLite Database
db = ConnectDB( os.path.join('Data', '02_processed', 'RbCPP', 'CrsDemHist.db') )
df.to_sql('FinalSnapshot', db.connection, if_exists='replace', index=False)
df
```

```python
df = db_EnrollmentFinalStatus.runQuery(""" --sql
    SELECT stu_prog_desc, stu_visa, reg_status, stu_admit_term_desc, reg_term_desc, crs, COUNT(*) AS demand
    FROM EnrollmentFinalStatus
    GROUP BY
        stu_admit_term_desc, reg_term_desc, crs
    ORDER BY
        stu_admit_term_code, reg_term_code, crs
""")
df.to_csv( os.path.join('Data', '02_processed', 'RbCPP', 'EnrollmentFinalStatus.csv'), index=False)
df.to_pickle( os.path.join('Data', '02_processed', 'RbCPP', 'EnrollmentFinalStatus.pkl'))
# Exporting to SQLite Database
df.to_sql('EnrollmentFinalStatus', db.connection, if_exists='replace', index=False)
df
```
