# Initialization
This file connects to the un-normalized Enrollment Data and performs some basic EDA on the data.

```python
# importing the required libraries
import sqlite3
import os
import pandas as pd
```

## Establishing connection to the SQL database
Using enrollment4EDA.db as the database file

```python
# configuring the sql connection and query execution
_       = os.sep
conn    = sqlite3.connect(".."+_+".."+_+".."+_+"Data"+_+"EnrollmentData"+_+"02_intermediate"+_+"enrollment4EDA.db")
cur     = conn.cursor()

def runQuery(query, commiting=False):
    cur.execute(query)
    if commiting:
        conn.commit()
    return pd.DataFrame(cur.fetchall(), columns=[i[0] for i in cur.description])
```

```python
# example test query
runQuery("""
    SELECT rec_id, stu_id
    FROM enrollment4EDA
    LIMIT 5;
""")
```

# EDA

```python
from IPython.display import display, HTML

display(HTML(runQuery("""
    SELECT reg_term_code, reg_status, rec_ext_date, SUM(rec_id)
    FROM enrollment4EDA
    WHERE
        reg_status IN ('**Registered**', '**Web Registered**', 'Wait Listed')
        AND
        EXTRACT(MONTH FROM rec_ext_date) = 02
        AND
        EXTRACT(DAY FROM rec_ext_date) = 15
    GROUP BY reg_term_code, reg_status, rec_ext_date;
""").to_html()))
```

```python
# get the number of students who have registered for a course in a given term on rec_ext_date Spetember 15th and October 15th
runQuery("""
    SELECT reg_term_code, reg_status, rec_ext_date, COUNT(DISTINCT stu_id)
    FROM enrollment4EDA
    WHERE
        reg_status IN ('**Registered**', '**Web Registered**', 'Wait Listed')
        AND
        rec_ext_date IN (
            '2017-09-15 00:00:00',
            '2018-02-15 00:00:00',
            '2018-09-15 00:00:00',
            '2019-02-15 00:00:00',
            '2019-09-15 00:00:00',
            '2020-02-15 00:00:00',
            '2020-09-15 00:00:00',
            '2021-02-15 00:00:00',
            '2021-09-15 00:00:00',
            '2022-02-15 00:00:00',
            '2022-09-15 00:00:00',
            '2023-02-15 00:00:00'
        )
    GROUP BY reg_term_code, reg_status, rec_ext_date;
""")
```

```python
# get the number of students who have registered for a course in a given term on rec_ext_date Spetember 15th and October 15th
runQuery("""
    SELECT reg_term_code, reg_status, rec_ext_date
    FROM enrollment4EDA
    WHERE
        reg_status IN ('**Registered**', '**Web Registered**', 'Wait Listed')
        AND
        YEAR(rec_ext_date) = 2023
    GROUP BY reg_term_code, rec_ext_date, reg_status;
""")
```

```python
runQuery("""
    SELECT strftime('%m-%d', rec_ext_date) AS days, rec_ext_date
    FROM enrollment4EDA
    ;
""")
```
